```python
#減少訓練資料
df =df.drop('link', axis=1).reset_index(drop=True)
```

```python
df.shape
```

```python
#Text Cleaning/ Text Preprocessing

# Step 1: Convert all text in the 'text' column to lowercase
# This is a common text preprocessing step to ensure case insensitivity
df['text'] = df['text'].str.lower()

# Step 2: Replace occurrences at the start of the string where there is a single word character followed by a whitespace
# The regular expression r'^\w\s' is used here
# '^' denotes the start of the string, '\w' is for any word character, and '\s' is for a whitespace
# This might be intended to remove certain patterns from the start of the string, but it's not very clear
df['text'] = df['text'].replace(r'^\w\s', ' ')

# Step 3: Replace newline characters ('\n') with a space
# This is done to convert multiline text into a single line, making it easier for certain types of text processing
# 'regex=True' indicates that the replacement is done using regular expressions
df['text'] = df['text'].replace(r'\n', ' ', regex=True)
```

```python
import torch
torch.cuda.is_available()
```

```python
import pandas as pd
from transformers import pipeline, BertTokenizer

# 初始化模型和分词器
model_name = 'bhadresh-savani/Distilbert-base-uncased-emotion'

# 指定使用GPU（如果可用）
device = 0 if torch.cuda.is_available() else -1

classifier = pipeline("text-classification", model=model_name, return_all_scores=True, device=device)
tokenizer = BertTokenizer.from_pretrained(model_name)
```

```python
# 为每种情绪创建一个新列
emotions = ['sadness', 'joy', 'love', 'anger', 'fear', 'surprise']
for emotion in emotions:
    df[emotion] = None

# 遍历 DataFrame，并应用模型
for index, row in df.iterrows():
    # 截断文本以适应模型的最大长度
    tokens = tokenizer.encode(row['text'], add_special_tokens=True, max_length=510, truncation=True)
    truncated_text = tokenizer.decode(tokens)
    predictions = classifier(truncated_text)

    for prediction in predictions[0]:
        df.at[index, prediction['label']] = prediction['score']

# 查看更新后的 DataFrame
df
```
